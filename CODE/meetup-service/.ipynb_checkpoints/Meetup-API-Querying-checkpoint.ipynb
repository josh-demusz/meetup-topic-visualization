```python
#Get a bunch of groups in San Fran (page=# of groups)
groups = requests.get('https://api.meetup.com/find/groups?&sign=true&photo-host=public&location=san+francisco&order=members&page=20',
                    headers=headers)
```

```python
location = 'san+francisco'
number = 1000

#Grab a number of groups based on the location and amount provided
def getGroups(location, amount, headers):
    geturl = 'https://api.meetup.com/find/groups?&sign=true&photo-host=public&location='+location+'&order=members&page='+str(amount)
    groups = requests.get(geturl, headers=headers)
    #Only a single request so no delay needed
    return groups.json()

#Retrieve all of the groups
groups = getGroups(location, number, headers)
```

```python
#Create function to reduce the amount of info in a group's json based on desired attributes
attributes = ['id', 'name', 'urlname', 'members', 'city', 'localized_location', 'region2', 'organizer', 
              'category']

#Reduce group to only the requested attributes
def reduceGroup(jsonentry):
    reducedgroup = dict((k, jsonentry[k]) for k in attributes)
    return reducedgroup

#Can apply this looped over all entries
reducedgroups = [reduceGroup(e) for e in groups]

#And grab each of their unique IDs
groupids = [v['id'] for v in reducedgroups]
```

```python
#Get all the events for a given group's URL (pulled from the json info)
def getEvents(groupid, headers):
    groupurl = [entry for entry in reducedgroups if (entry['id'] == groupid)][0]['urlname']
    url = 'https://api.meetup.com/'+groupurl+'/events?&sign=true&photo-host=public&page=20'
    resp = requests.get(url, headers=headers)
    #Implement connection check here?
    print(resp)
    events = [entry['id'] for entry in resp.json()]
    return events
```

```python
#Build dict based on the groupids
groupeventdict = dict.fromkeys(groupids)

#Fill the dictionary for each groupid with a list of their events
i = 0 
for gid in groupids:
    events = getEvents(gid, headers)
    groupeventdict[gid] = events
    print(i)
    i += 1
    #Prevent being disconnected due to throttling by the API
    time.sleep(0.1)
```

```python
#Next steps: Create event comment puller, also want topics/categories, 
#Get info for large amount of groups located in San Francisco
#Grab the events for each group
#Collect the comments for them all
#Put the data into SQLite DB
```

```python
#Grab comments for an event, returns json object
def getComments(groupurl, eventid, headers):
    url = 'https://api.meetup.com/'+groupurl+'/events/'+eventid+'/comments?&sign=true&photo-host=public&page=100'
    resp = requests.get(url, headers=headers)
    #Implement connection check here
    return resp.json()
```

```python
#Cleaning up the groupeventdict in case not all groups were grabbed correctly
newdict = dict((k, groupeventdict[k]) for k in groupeventdict.keys() if type(groupeventdict[k]) == list)

#Make a tuple of group id and event id for every event for a given group
eventgrouptuples = [(k,i) for k,v in newdict.items() for i in v]

#Split the tuples into two lists to be inserted into the df 
eventlist = [tup[1] for tup in eventgrouptuples]
grouplist = [tup[0] for tup in eventgrouptuples]

#Create a corresponding list of groupurls for each event entry
groupurls = []
for gid in grouplist:
    gurl = [entry['urlname'] for entry in groups if entry['id']==gid][0]
    groupurls.append(gurl)
```

```python
#Create base dataframe based on eventids
edf = pd.DataFrame()
edf['EventID'] = eventlist
edf['GroupID'] = grouplist
edf['GroupURL'] = groupurls
```

```python
print('Retrieved Events: ', len(edf))
edf.sample(25)
```

```python
#Get the comments and build a list of json results for every event in the df
comments = []      
for index, row in edf.iterrows():
    eventcomments = getComments(row['GroupURL'], row['EventID'], headers)
    comments.append(eventcomments)
    time.sleep(0.1)
```

```python
#Extract a list of the actual comments
def extractComments(jsonstuff):
    #If the result is a dict then a privacy error was returned, fill the comment list with NA
    if type(jsonstuff) == dict:
        comments = ['NA']
        return comments
    #If the result is a list then we did retrieve the comments, even if it's an empty list
    if type(jsonstuff) == list:
        comments = [e['comment'] for e in jsonstuff]
        return comments
```

```python
#Create a new column of all the eventjson objects retrieved
edf['EventCommentsJSON'] = comments

#Apply the comment extractor to all of the rows
edf['Comments'] = edf['EventCommentsJSON'].apply(extractComments)
```
